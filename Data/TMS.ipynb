```python
def tata_val(y_true, y_pred):
    return np.mean((np.abs(y_pred - y_true) * 200/ (np.abs(y_pred) + np.abs(y_true))))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
def rmse(y_true, y_pred):
    rmse_t = mean_squared_error(y_true, y_pred)
    return rmse_t
```

### Augumented Dickey - Fuller test

#### Null hypothesis - The series is not stationary
#### Alternate hypothesis - The series is stationary

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(test_data['Close'])
print('ADF Statistics: %f' %result[0])
print('p-value:%f' %result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' %(key,value))
```

### Here we see that p - value is more than 0.05. Hence our alternate hypothesis will be rejected . Thus this series is non-stationary.

## LSTM

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
T_Close
```

```python
train_T = T_Close.iloc[:2300]
test_T = T_Close.iloc[2300:]
```

```python
test_T = test_T[0:184]
```

```python
test_T.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train_T)
```

```python
scaler_train = scaler.transform(train_T)
scaler_test = scaler.transform(test_T)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
len(scaler_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
n_input = 184
n_feature = 1

train_generator_T = TimeseriesGenerator(scaler_train, scaler_train, length = n_input, batch_size = 1)
```

```python
models = Sequential()

models.add(LSTM(128, activation = 'relu', input_shape=(n_input, n_feature), return_sequences = True))
models.add(LSTM(128, activation = 'relu', input_shape=(n_input, n_feature), return_sequences = True))
models.add(LSTM(128, activation = 'relu', input_shape=(n_input, n_feature), return_sequences = False))
models.add(Dense(1))
```

```python
models.summary()
```

```python
models.compile(optimizer = 'adam', loss= 'mean_squared_error')
models.fit_generator(train_generator_T, epochs = 5)
```

```python
losses_lstm = models.history.history['loss']
plt.figure(figsize = (12,4))
plt.xticks(np.arange(0,21,1))
plt.plot(range(len(losses_lstm)), losses_lstm)
```

```python
lstm_predictions_scaled = list()

batch = scaler_train[-n_input:]
current_batch = batch.reshape((1, n_input, n_feature))

for i in range(len(test_T)):   
    lstm_pred = models.predict(current_batch)[0]
    lstm_predictions_scaled.append(lstm_pred) 
    current_batch = np.append(current_batch[:,1:,:],[[lstm_pred]],axis=1)
```

```python
lstm_predictions_scaled
```

```python
lstm_predictions = scaler.inverse_transform(lstm_predictions_scaled)
lstm_predictions
```

```python
test_T['LSTM_Predictions'] = lstm_predictions
test_T
```

```python
test_T['Close'].plot(figsize=(20,10), legend = True)
test_T['LSTM_Predictions'].plot(legend = True)
```

```python
from statsmodels.tools.eval_measures import rmse
lstm_rmse_error = rmse(test_T['Close'], test_T["LSTM_Predictions"])
lstm_mse_error = lstm_rmse_error**2
mean_value = T_Close['Close'].mean()

print(f'MSE Error: {lstm_mse_error}\nRMSE Error: {lstm_rmse_error}\nMean: {mean_value}')
```
